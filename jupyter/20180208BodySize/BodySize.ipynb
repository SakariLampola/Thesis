```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid", color_codes=True)
def plot(item, mu, sigma):
    s = np.random.lognormal(mu, sigma, 10000)
    label = item + " mu=" + str(round(mu,2)) + " sigma= " + str(round(sigma,2))
    sns.distplot(s, label=label)
    plt.legend()
    
plt.figure(), plot("Aeroplane", np.log(20), 0.35)
plt.figure(), plot("Bicycle", np.log(1.5), 0.04)
plt.figure(), plot("Bird", np.log(0.15), 0.35)
plt.figure(), plot("Boat", np.log(6.0), 0.35)
```

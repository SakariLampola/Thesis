```python
mu = np.array([m_x, m_y, m_z, m_vx, m_vy, m_vz])
```

```python
mu # mean for location and velocity
```

```python
sigma = np.array([[s_x_x,  s_x_y,  s_x_z,  s_x_vx,  s_x_vy,  s_x_vz],
                  [s_y_x,  s_y_y,  s_y_z,  s_y_vx,  s_y_vy,  s_y_vz],
                  [s_z_x,  s_z_y,  s_z_z,  s_z_vx,  s_z_vy,  s_z_vz],
                  [s_vx_x, s_vx_y, s_vx_z, s_vx_vx, s_vx_vy, s_vx_vz],
                  [s_vy_x, s_vy_y, s_vy_z, s_vy_vx, s_vy_vy, s_vy_vz],
                  [s_vz_x, s_vz_y, s_vz_z, s_vz_vx, s_vz_vy, s_vz_vz]])
```

```python
sigma # covariance for location and velocity
```

```python
samples = np.random.multivariate_normal(mu, sigma,1000)
```

```python
samples
```

```python
x=samples[:,0]
y=samples[:,1]
z=samples[:,2]
```

```python
sns.regplot(x=x, y=z, fit_reg=False)
```

```python
vx=samples[:,3]
vy=samples[:,4]
vz=samples[:,5]
```

```python
sns.regplot(x=vx, y=vz, fit_reg=False)
```

```python
P = samples[:,0:3]
P
```

```python
V = samples[:,3:]
V
```

```python
P.shape
```

```python
V.shape[0]
```

```python
t = np.zeros((P.shape[0],1))
```

```python
for i in range(P.shape[0]):
    t[i] = P[i,:].dot(V[i,:].T)/pow(np.linalg.norm(V[i,:]),2)
```

```python
t = -t # Math has to checked.... why this minus sign?
```

```python
t.shape
```

```python
t
```

```python
x_end = np.zeros((t.shape[0],1))
y_end = np.zeros((t.shape[0],1))
z_end = np.zeros((t.shape[0],1))
```

```python
for i in range(t.shape[0]):
    x_end[i] = x[i] + t[i]*vx[i]
    y_end[i] = y[i] + t[i]*vy[i]
    z_end[i] = z[i] + t[i]*vz[i]
```

```python
ri = np.random.lognormal(1.28, 0.25, t.shape[0])/2 # car body sizes
ro = np.random.lognormal(0.51, 0.07, t.shape[0])/2 # observer (person) body sizes
xo = np.zeros((P.shape[0],1))
zo = np.zeros((P.shape[0],1))
```

```python
sns.regplot(x=x, y=z, fit_reg=False) # Blue, starting point
sns.regplot(x=x_end, y=z_end, fit_reg=False) # Green, nearest point 
```

```python
area=200*ri*ri
plt.scatter(x,z,s=area,alpha=0.3)
plt.scatter(x_end,z_end,s=area,alpha=0.3)
area=200*ro*ro
plt.scatter(xo,zo,s=area,alpha=0.3)
```

```python
p_collision = 0.0
for i in range(P.shape[0]):
    if np.linalg.norm(np.array([x_end[i],y_end[i],z_end[i]])) < (ri[i]+ro[i]):
                      p_collision +=1
p_collision /= P.shape[0]
p_collision
```

```python
plt.scatter(x,z,alpha=1.0,s=2) # blue, start point
plt.scatter(x_end,z_end,alpha=1.0,s=2) # green, end point
plt.scatter(xo,zo,alpha=1.0,s=220) # red, observer
```

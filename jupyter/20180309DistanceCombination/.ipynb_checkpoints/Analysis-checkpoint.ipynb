```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
```

```python
ed=np.linspace(0,100,num=100)
```

```python
ext = 20.0
stereo=np.exp(-ed/ext)
size=1-stereo
```

```python
plt.plot(ed,stereo,label="k,stereo")
plt.plot(ed,size,label="k,size")
plt.legend()
```

```python
dstereo = 15.0
dsize = 21.0
```

```python
d = dstereo*stereo+dsize*size
```

```python
plt.plot(ed,d)
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
import cv2
```

```python
im_left = cv2.imread('left.png',0)
im_right = cv2.imread('right.png',0)
```

```python
plt.imshow(im_left)
```

```python
plt.imshow(im_right)
```

```python
stereo = cv2.StereoBM_create(numDisparities=6*16, blockSize=27)
disparity = stereo.compute(im_left,im_right)
minValue = disparity.min()
maxValue = disparity.max()
disp_rgb = np.uint8(255 * (disparity - minValue) / (maxValue - minValue))
disp = cv2.applyColorMap(disp_rgb, cv2.COLORMAP_WINTER)
plt.imshow(disp)
```

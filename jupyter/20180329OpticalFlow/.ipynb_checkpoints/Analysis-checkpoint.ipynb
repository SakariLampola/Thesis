```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
import cv2
mpl.rcParams['figure.figsize'] = [16, 4]
```

```python
im_1 = cv2.imread('1.png',0)
im_2 = cv2.imread('2.png',0)
plt.imshow(im_1)
```

```python
plt.imshow(im_2)
```

```python
flow = cv2.calcOpticalFlowFarneback(im_1,im_2, None, 0.5, 3, 15, 3, 5, 1.2, 0)
```

```python
flow_x=flow[...,0]
```

```python
plt.hist(flow_x.flatten())
```

```python
minValue = flow_x.min()
maxValue = flow_x.max()
flow_gray = np.uint8(255.0 * (flow_x - minValue) / (maxValue - minValue))
meanValue = flow_x.mean()
meanValue
```

```python
plt.hist(flow_gray.flatten())
```

```python
plt.imshow(flow_gray)
```

```python
disp = cv2.applyColorMap(flow_gray, cv2.COLORMAP_JET)
plt.imshow(disp)
```
